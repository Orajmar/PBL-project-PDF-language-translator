```python
%%writefile app.py
import os
import sys
import pandas as pd
import numpy as np
import streamlit as st
sys.path.insert(0,'/content/drive/MyDrive/PDFTranslator/')
from reportlab.lib.styles import ParagraphStyle, getSampleStyleSheet
from reportlab.pdfbase import pdfmetrics
from reportlab.pdfbase.ttfonts import TTFont

from PyPDF2 import PdfReader
from googletrans import Translator, LANGUAGES
from reportlab.lib.pagesizes import letter
from reportlab.lib.styles import ParagraphStyle
from reportlab.platypus import Paragraph
from reportlab.platypus import SimpleDocTemplate

URL_COM = 'translate.googleapis.com'


directory = "/content/drive/MyDrive/PDFTranslator/fonts"

BODY_FONT_SIZE = 12
TITLE_FONT_SIZE = 14
ALIGNMENT_RIGHT = 0
SPACING_BETWEEN_LINES = 20
SPACE_AFTER = 14
FIRST_LINE_INDENT = 12
REGULAR_HIN_NAME = 'KrutiDev010'
REGULAR_FONT_NAME = 'NotoSerif-Regular'
BOLD_FONT_NAME = 'NotoSerif-Bold'
ITALIC_FONT_NAME = 'NotoSerif-Italic'
BOLDITALIC_FONT_NAME = 'NotoSerif-BoldItalic'

REGULAR_FONT_FILE = os.path.join(directory, REGULAR_FONT_NAME + ".ttf")
BOLD_FONT_FILE = os.path.join(directory, BOLD_FONT_NAME + ".ttf")
ITALIC_FONT_FILE = os.path.join(directory, ITALIC_FONT_NAME + ".ttf")
BOLDITALIC_FONT_FILE = os.path.join(directory, BOLDITALIC_FONT_NAME + ".ttf")

pdfmetrics.registerFont(TTFont('NotoSerif-Regular', REGULAR_FONT_FILE))
pdfmetrics.registerFont(TTFont('NotoSerif-BoldItalic', BOLDITALIC_FONT_FILE))
pdfmetrics.registerFont(TTFont('NotoSerif-Bold', BOLD_FONT_FILE))
pdfmetrics.registerFont(TTFont('NotoSerif-Italic', ITALIC_FONT_FILE))

style = getSampleStyleSheet()


def create_body_font(name):
    return ParagraphStyle(name,
                          fontName=name,
                          fontSize=BODY_FONT_SIZE,
                          parent=style['Normal'],
                          alignment=ALIGNMENT_RIGHT,
                          leading=SPACING_BETWEEN_LINES,
                          firstLineIndent=FIRST_LINE_INDENT,
                          spaceAfter=SPACE_AFTER)


def create_title_font(name, style_name):
    return ParagraphStyle(name,
                          fontName=name,
                          fontSize=TITLE_FONT_SIZE,
                          parent=style[style_name],
                          alignment=ALIGNMENT_RIGHT,
                          leading=SPACING_BETWEEN_LINES,
                          firstLineIndent=FIRST_LINE_INDENT,
                          spaceAfter=SPACE_AFTER)


regular = create_body_font(REGULAR_FONT_NAME)
regularhin = create_body_font(REGULAR_HIN_NAME)
bold = create_body_font(BOLD_FONT_NAME)
italic = create_body_font(ITALIC_FONT_NAME)
boldItalic = create_body_font(BOLDITALIC_FONT_NAME)
heading1 = create_title_font(BOLD_FONT_NAME, 'Heading1')
heading2 = create_title_font(BOLD_FONT_NAME, 'Heading2')
heading3 = create_title_font(BOLD_FONT_NAME, 'Heading3')
heading4 = create_title_font(BOLD_FONT_NAME, 'Heading4')
heading5 = create_title_font(BOLD_FONT_NAME, 'Heading5')



def get_translated_page_content(reader, lang):
    num_pages = len(reader.pages)
    page_contents = []
    translator = Translator(service_urls=[URL_COM])
    for p in range(num_pages):
        page = reader.pages[p]
        text = page.extract_text()
        translation = translator.translate(text, dest=lang)
        result_text = translation.text.splitlines()
        #print(result_text)
        page_contents.extend(result_text)
    return page_contents

def translate_pdf(path, lang):
    file = open(path, 'rb')
    reader = PdfReader(file)
    page_contents = get_translated_page_content(reader, lang)
    #print(page_contents)

    page_text = []
    pdf = SimpleDocTemplate(fname, title=pdf_file.name[:-4], pagesize=letter)
    for text in page_contents:
        page_text.append(Paragraph(text, encoding='utf-8', style=regular))
    pdf.build(page_text)

st.set_page_config(page_title="Online PDF Language Translator", page_icon="")
style_hide = '''
<style>
h1, h3{
  color:white;
}

.uploadedFileName{
  color:white;
}

.stApp {
  background-image: url("https://cdn.pixabay.com/photo/2018/04/28/16/27/energy-3357613_960_720.jpg");
  background-attachment: fixed;
  background-size: cover;
    }

#MainMenu {visibility: hidden;}
header, footer {visibility: hidden;}

button[title="View fullscreen"]{
    visibility: hidden;
    }

section[data-testid="stSidebar"] div:first-child {
        top: 0;
        height: 100vh;
    }
.css-eczf16 {display: none}

.div{
  padding-top:0px;
}

.stMarkdown{
  color:white;
}

.css-rvekum {
    color: white;
    font-size: 1.2rem;
}

.uploadedFile{
    background-color: white;
    border-radius: 3px;
    margin-top:15px;
}

.uploadedFileName{
    color: #31333f;
}

tr{
  color: white;
}

thead tr th:first-child {display:none}
tbody th {display:none}
thead {display:none}

</style>
'''

st.markdown(style_hide, unsafe_allow_html=True)

# st.image("/content/top.png")

st.title("Online PDF Language Translator")
st.subheader("Choose PDF File")
pdf_file = st.file_uploader("",type=["pdf"], label_visibility="hidden")

key_list = list(LANGUAGES.keys())
val_list = list(LANGUAGES.values())

if pdf_file is not None:
	# To See details
	st.subheader("Select Language")
	selected_lang = st.selectbox(label="", options=val_list, label_visibility="hidden")
	position = val_list.index(selected_lang)
	selected_lang = key_list[position]

	file_details = {"File Name":pdf_file.name, "Size":pdf_file.size}
	st.write(file_details)

	with open(os.path.join("/content",pdf_file.name),"wb") as f:
		f.write((pdf_file).getbuffer())

	st.write("")
	st.caption("File Uploaded")
	fname = f'{selected_lang}_{pdf_file.name}'

	st.write("")
	rst = st.button("Translate",on_click=translate_pdf(pdf_file.name, selected_lang))
	st.write("")
	st.write("")

	if rst:
		with open(fname, "rb") as pdf_file:
			PDFbyte = pdf_file.read()
		st.download_button(
		label="Download",
    data=PDFbyte,
    file_name=fname,
    mime='application/pdf'
    )


my_array = np.array(list(val.capitalize() for val in val_list))
my_array = np.append(my_array,"")
my_array = my_array.reshape(4, 27).T
df = pd.DataFrame(my_array, columns=('col %d' % i for i in range(4)))
st.write("")
st.subheader("Languages Supported")
st.table(df)

st.subheader("Frequently Asked Questions")
st.markdown('''
**Is it free to translate a PDF into English?**
>Yes, you can translate a PDF document from any language to English completely free of charge.

''')

st.markdown('''
**How can I translate a document into English?**
>Simply drag and drop your document into the form, the online document translator will detect the document’s original language, then select "English" as the language you want it translated into, then just click the "Translate" button.

''')
```
